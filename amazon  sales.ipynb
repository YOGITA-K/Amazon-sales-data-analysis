```python
#Calculated total revenue, total cost, and total profit.
total_revenue = df['Total Revenue'].sum()
total_cost = df['Total Cost'].sum()
total_profit = df['Total Profit'].sum()

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Cost: ${total_cost:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
```

```python
#Profitability by Region
profit_by_region = df.groupby('Region')['Total Profit'].sum().reset_index()
print(profit_by_region)
```

```python
# Calculating  delivery lead time
df['Delivery Lead Time'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
#  the updated DataFrame with delivery lead time
print(df[['Order Date', 'Ship Date', 'Delivery Lead Time']])
```

```python
# Checking  for rows with NaT in Order Date or Ship Date
missing_dates = df[df['Order Date'].isna() | df['Ship Date'].isna()]
```

```python
# Displaying  rows with missing dates
print("Rows with missing Order Date or Ship Date:")
print(missing_dates)
```

```python
# Optionally, filter out rows with NaT in Delivery Lead Time
df_filtered = df.dropna(subset=['Delivery Lead Time'])
```

```python
# Display the updated DataFrame with delivery lead time
print("\nUpdated DataFrame with Delivery Lead Time:")
print(df_filtered[['Order Date', 'Ship Date', 'Delivery Lead Time']])
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
# Aggregating Total Revenue by Year
revenue_trend = df.groupby('Year')['Total Revenue'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(revenue_trend['Year'], revenue_trend['Total Revenue'], marker='o')
plt.title('Revenue Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.xticks(revenue_trend['Year'])
plt.grid()
plt.show()
```

```python
# Extract the year and month from 'Order Date'
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
# Group the data by year and month and calculate total revenue, total cost, total profit, and total units sold
trends_monthwise = df.groupby(['Year', 'Month']).agg(
    Total_Revenue=('Total Revenue', 'sum'),
    Total_Cost=('Total Cost', 'sum'),
    Total_Profit=('Total Profit', 'sum'),
    Total_Units_Sold=('Units Sold', 'sum')
).reset_index()
```

```python
# Create a 'Month-Year' column for easier plotting
trends_monthwise['Month-Year'] = trends_monthwise['Year'].astype(str) + '-' + trends_monthwise['Month'].astype(str).str.zfill(2)
```

```python
# Print the month-wise trends
print(trends_monthwise[['Month-Year', 'Total_Revenue', 'Total_Cost', 'Total_Profit', 'Total_Units_Sold']])
```

```python
# Visualizing the trends
plt.figure(figsize=(12, 6))
plt.plot(trends_monthwise['Month-Year'], trends_monthwise['Total_Revenue'], marker='o', label='Total Revenue')
plt.plot(trends_monthwise['Month-Year'], trends_monthwise['Total_Profit'], marker='o', label='Total Profit')
plt.plot(trends_monthwise['Month-Year'], trends_monthwise['Total_Cost'], marker='o', label='Total Cost')
plt.title('Sales Trends by Month')
plt.xlabel('Month-Year')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45)  # Rotate x labels for better visibility
plt.legend()
plt.grid()
plt.tight_layout()  # Adjust layout to make room for rotated labels
plt.show()
```

```python
# Group the data by year and calculate total revenue, total cost, total profit, and total units sold
trends_yearwise = df.groupby('Year').agg(
    Total_Revenue=('Total Revenue', 'sum'),
    Total_Cost=('Total Cost', 'sum'),
    Total_Profit=('Total Profit', 'sum'),
    Total_Units_Sold=('Units Sold', 'sum')
).reset_index()
```

```python
# Visualizing the trends
plt.figure(figsize=(10, 6))
plt.plot(trends_yearwise['Year'], trends_yearwise['Total_Revenue'], marker='o', label='Total Revenue')
plt.plot(trends_yearwise['Year'], trends_yearwise['Total_Profit'], marker='o', label='Total Profit')
plt.plot(trends_yearwise['Year'], trends_yearwise['Total_Cost'], marker='o', label='Total Cost')
plt.title('Sales Trends by Year')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.xticks(trends_yearwise['Year'])  # Show all years on the x-axis
plt.legend()
plt.grid()
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
# Group the data by year and month and calculate total revenue, total cost, total profit, and total units sold
trends_yearly_monthwise = df.groupby(['Year', 'Month']).agg(
    Total_Revenue=('Total Revenue', 'sum'),
    Total_Cost=('Total Cost', 'sum'),
    Total_Profit=('Total Profit', 'sum'),
    Total_Units_Sold=('Units Sold', 'sum')
).reset_index()

```

```python
# Create a 'Month-Year' column for easier plotting
trends_yearly_monthwise['Month-Year'] = trends_yearly_monthwise['Year'].astype(str) + '-' + trends_yearly_monthwise['Month'].astype(str).str.zfill(2)
```

```python
# Visualizing the trends
plt.figure(figsize=(12, 6))

# Plot each year in a different color
for year in trends_yearly_monthwise['Year'].unique():
    monthly_data = trends_yearly_monthwise[trends_yearly_monthwise['Year'] == year]
    plt.plot(monthly_data['Month-Year'], monthly_data['Total_Revenue'], marker='o', label=f'Total Revenue {year}')
    plt.plot(monthly_data['Month-Year'], monthly_data['Total_Profit'], marker='o', linestyle='--', label=f'Total Profit {year}')
    plt.plot(monthly_data['Month-Year'], monthly_data['Total_Cost'], marker='o', linestyle=':', label=f'Total Cost {year}')

plt.title('Yearly Month-wise Sales Trends')
plt.xlabel('Month-Year')
plt.ylabel('Amount ($)')
plt.xticks(rotation=45)  # Rotate x labels for better visibility
plt.legend()
plt.grid()
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
# Group the data by Sales Channel and calculate total revenue, total cost, total profit, and total units sold
trends_by_channel = df.groupby('Sales Channel').agg(
    Total_Revenue=('Total Revenue', 'sum'),
    Total_Cost=('Total Cost', 'sum'),
    Total_Profit=('Total Profit', 'sum'),
    Total_Units_Sold=('Units Sold', 'sum')
).reset_index()
```

```python
# Visualizing the trends by sales channel
plt.figure(figsize=(8, 5))
```

```python
# Bar plot for Total Revenue, Total Profit, and Total Cost
bar_width = 0.25
index = range(len(trends_by_channel))
```

```python
# Creating bar plots
plt.bar(index, trends_by_channel['Total_Revenue'], bar_width, label='Total Revenue', color='b')
plt.bar([i + bar_width for i in index], trends_by_channel['Total_Profit'], bar_width, label='Total Profit', color='g')
plt.bar([i + 2 * bar_width for i in index], trends_by_channel['Total_Cost'], bar_width, label='Total Cost', color='r')
```
